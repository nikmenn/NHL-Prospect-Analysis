```python
#Reading in the 2015 draft and their NHL stats (from HockeyReference)
nhl_2015_df = pd.read_csv('HR_2015_NHL_Draft.csv')
nhl_2015_df.head()
```

```python
#Removing the unecessary information from the Player Column (part 1)
nhl_2015_df['Player'] = nhl_2015_df['Player'].str.lower()
nhl_2015_df['Player'] = nhl_2015_df['Player'].str.split('\\')
nhl_2015_df['Player_temp'] = nhl_2015_df['Player'].str[0]
nhl_2015_df.head()
```

```python
#Removing the unecessary information from the Player Column (part 2)
nhl_2015_df['Player'] = nhl_2015_df['Player_temp']
nhl_2015_df = nhl_2015_df.drop(['Player_temp'], axis=1)
nhl_2015_df['Player'] = nhl_2015_df['Player'].str.title()
nhl_2015_df.head()
```

```python
#Rename 'Player' column to match the column name from the OHL dataframe
nhl_2015_df = nhl_2015_df.rename(columns={"Player": "Name"})
nhl_2015_df.head()
```

```python
#Creating a new dataframe that just consists of certain columns from the ohl_2015_df
ohl_2015_forwards = pd.DataFrame()
ohl_2015_forwards['Name'] = ohlf_2015_df['NAME']
ohl_2015_forwards['OHL_Points'] = ohlf_2015_df['TP']
ohl_2015_forwards['OHL_PointsPG'] = ohlf_2015_df['TP/GP']
ohl_2015_forwards['OHL_GP'] = ohlf_2015_df['GP']
ohl_2015_forwards.head()
```

```python
nhl_2015_df.columns
```

```python
#dropping unecessary columns from the nhl_2015_df 
nhl_2015_df = nhl_2015_df.drop(['To', 'Age','PIM', 'GP.1', 'W', 'L', 'T/O', 'SV%',
       'GAA', 'PS'], axis=1)
nhl_2015_df.head()
```

```python
#Merging the nhl_2015_df dataset with the ohl_2015_forwards dataset
#Using the default inner merge so only the ones with that are present in both tables will be in the resulting table

nhl_2015_forwards = pd.merge(nhl_2015_df, ohl_2015_forwards)
nhl_2015_forwards.head()
```

```python
#Creating a new column for NHL points per game 
nhl_2015_forwards['NHL_PointsPG'] = nhl_2015_forwards['PTS']/nhl_2015_forwards['GP']
nhl_2015_forwards.head()
```

```python
#Creating a Bokeh plot of some of the above data
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

output_file("2015_NHL_Draft_Forwards.html", title="2015 NHL Draft Forwards")

TOOLTIPS = [
    ("Name","@Name"),
    ("Team", "@Team"),
    ("OHL PPG", "@OHL_PointsPG"),
    ("NHL PPG", "@NHL_PointsPG")
]

#Create the figure and glyphs(circles)
fig = figure(tooltips = TOOLTIPS)
p = fig.circle(x='OHL_PointsPG', y='NHL_PointsPG', source = nhl_2015_forwards)

#Adding the axis labels and coloring/altering the design of the glyphs
fig.xaxis.axis_label = 'OHL Points Per Game'
fig.yaxis.axis_label = 'NHL Points Per Game'
glyph = p.glyph
glyph.size = 15
glyph.fill_color = 'red'
glyph.line_color = 'black'

show(fig)
```
