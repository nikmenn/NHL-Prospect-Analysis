```python
import pandas as pd
import numpy as np
```

```python
nhl_df = pd.read_csv('nhl_f_combined_junior_pro_stats.csv')
nhl_df.head()
```

```python
nhl_df.columns
```

```python
#drop unnecessary column
nhl_df = nhl_df.drop(
        ['Unnamed: 0'], axis=1)
nhl_df.head()
```

```python
#Changing nan values to 0
nhl_df = nhl_df.fillna(0)

#Only using players who have played at least 150 games in the NHL
nhl_df = nhl_df[nhl_df['GP'] >= 150]

nhl_df = nhl_df.reset_index(drop=True)
nhl_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split

from yellowbrick.datasets import load_concrete
from yellowbrick.regressor import ResidualsPlot, residuals_plot
```

```python
X = nhl_df[['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']].copy()
y = nhl_df['NHL_PointsPG']
```

```python
# Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#documentation from https://www.scikit-yb.org/en/latest/api/regressor/residuals.html
lr = LinearRegression()
lr_visualizer = ResidualsPlot(lr)
lr_visualizer.fit(X_train,y_train)
lr_visualizer.score(X_test, y_test)
lr_visualizer.show() 
```

```python
lr_prediction_frame = nhl_df.copy()
lr_prediction_frame['Predicted_NHLPointsPG'] = lr.predict(lr_prediction_frame
                                                       [['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']])
lr_prediction_frame[['Name','Predicted_NHLPointsPG','NHL_PointsPG']][:25]
```

```python
# Create and train the Support Vector Machine  (Regressor)
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_visualizer = ResidualsPlot(svr_rbf)
svr_visualizer.fit(X_train,y_train)
svr_visualizer.score(X_test, y_test)
svr_visualizer.show() 
```

```python
# RandomForest Regression model
from sklearn.ensemble import RandomForestRegressor
#Faster way to do the visualization using residuals_plot like this but you can also use ResidualsPlot like above
forest_model = RandomForestRegressor()
forest_visualizer = residuals_plot(forest_model, X_train, y_train, X_test, y_test)
```

```python
forest_prediction_frame = nhl_df.copy()
forest_prediction_frame['Predicted_NHLPointsPG'] = forest_model.predict(forest_prediction_frame
                                                       [['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']])
forest_prediction_frame[['Name','Predicted_NHLPointsPG','NHL_PointsPG']][:25]
```

```python
draft_2020 = pd.read_csv('chl_f_2020_trunc.csv')
draft_2020 = draft_2020.drop(
        ['Unnamed: 0'], axis=1)
draft_2020.head()
```

```python
draft_2020['Forest_Predicted_NHLPointsPG'] = forest_model.predict(draft_2020[
    ['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']])
draft_2020['SVR_Predicted_NHLPointsPG'] = svr_visualizer.predict(draft_2020[
    ['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']])
draft_2020['LR_Predicted_NHLPointsPG'] = lr_visualizer.predict(draft_2020[
    ['CHL_PointsPG','CHL_EV_P1/GP','CHL_EV_P1/60']])
```

```python
draft_2020_prediction = draft_2020[['Name', 'CHL_PointsPG', 'Forest_Predicted_NHLPointsPG', 
            'SVR_Predicted_NHLPointsPG', 'LR_Predicted_NHLPointsPG']]
draft_2020_prediction = draft_2020_prediction[draft_2020_prediction['CHL_PointsPG'] >= .3]
draft_2020_prediction[:25]
```

```python
#Creating a Bokeh plot of some of the above data
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

output_file("2020_NHL_Draft_Forwards_Predictions.html", title="2020 NHL Draft Forwards Predictions")

TOOLTIPS = [
    ("Name","@Name"),
    ("CHL PPG", "@CHL_PointsPG"),
    ("LR Predicted NHL PPG", "@LR_Predicted_NHLPointsPG"),
]

#Create the figure and glyphs(circles)
fig = figure(tooltips = TOOLTIPS)
lr_p = fig.circle(x='CHL_PointsPG', y='LR_Predicted_NHLPointsPG', source = draft_2020_prediction)
#forest_p = fig.circle(x='CHL_PointsPG', y='Forest_Predicted_NHLPointsPG', source = draft_2020_prediction)

#Adding the axis labels and coloring/altering the design of the glyphs
fig.xaxis.axis_label = 'CHL Points Per Game'
fig.yaxis.axis_label = 'Predicted NHL Points Per Game'
lr_glyph = lr_p.glyph
lr_glyph.size = 10
lr_glyph.fill_color = 'red'
lr_glyph.line_color = 'black'

show(fig)
```
